# Generate Indexes and Mask

The ImageProcessing generates NDVI and NDRE for all products of interest, and masks each polygon in the geopanda, saving files inside local directories for the polygon and specific index (specifying the product's date).

```python
gdf = FileProcessing().read_from_kml('plot_files/us_multi_plot.kml')
```

```python
processor = ImageProcessing(gdf,database=local_storage)
```

```python
folder, date = processor.get_folders(local_db)
```

```python
folder
```

```python
for key in folder.keys():
    sub_folders = folder[key]
    sub_dates = date[key]
    for i in range(len(sub_folders)):
        img_path = processor.get_img_path(sub_folders[i])
        processor.generate_ndvi(img_path)
        processor.generate_ndre(img_path)
        processor.mask_ndvi(img_path,sub_dates[i],key)
        processor.mask_ndre(img_path,sub_dates[i],key)
        processor.mask_tci(img_path,sub_dates[i],key)
```
